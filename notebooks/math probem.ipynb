```python
import math

first_circle = input().split()
second_circle = input().split()

first_circle = list(map(int, first_circle))
second_circle = list(map(int, second_circle))
dist = math.sqrt((first_circle[0] - second_circle[0])**2 + (first_circle[1] - second_circle[1])**2)
if dist == 0:
    print("Circles are overlapping")
elif dist < first_circle[2] + second_circle[2]:
    print("Yes")
else:
    print("No")
```
